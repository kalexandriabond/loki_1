```python
# import dabest
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = '/home/krista/Dropbox/volatileValues/simple_rt_experiment_probabilityC/analysis/aggregated_data/'
ls_obs_data = pd.read_csv(data_path+"combined_learning_sig_obs_data.csv")
```

```python
ls_obs_data.head()
```

```python
grouped_df = ls_obs_data.groupby(['ID', 'condition'])[['p_id_accuracy', 'rt']].mean()
grouped_df.head()
```

```python
n_boot=5000
```

```python
f_conf_acc, results_conf_acc = dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='p_id_accuracy', 
            color_col='condition', idx=['lc', 'hc'], n_boot=n_boot, paired=True, show_pairs=False,
            swarm_label='Accuracy',  swarm_ylim = (0.4, 1))
plt.axhline(linewidth=2, color='k', y = -0.22, ls='dashed', alpha=0.3)
plt.savefig('conflict_acc_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_vol_acc, results_vol_acc = dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='p_id_accuracy', 
            color_col='condition', idx=['lv', 'hv'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Accuracy', swarm_ylim = (0.4, 1))
plt.axhline(linewidth=2, color='k', y = -0.175, ls='dashed', alpha=0.3)

plt.savefig('vol_acc_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_vol_rt, results_vol_rt = dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='rt', 
            color_col='condition', idx=['lv', 'hv'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Reaction Time (s)', swarm_ylim = (0.24, 0.5))
plt.savefig('vol_rt_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
f_conf_rt, f_results_rt = dabest.plot(data=grouped_df, fig_size=(5,7),  x='condition', y='rt', 
            color_col='condition', idx=['lc', 'hc'], n_boot=n_boot, paired=True,show_pairs=False, 
            swarm_label='Reaction Time (s)', swarm_ylim = (0.25, 0.5))
plt.savefig('conflict_rt_estimation_plot.png', dpi=600, bbox_inches='tight', transparent=0)
```

```python
sns.set(style="white", palette="muted", color_codes=True, context='talk')
# sns.set(font_scale=1.8)  

lc_dist = sns.distplot(ls_obs_data.loc[ls_obs_data.condition == 'lc', 'rt'], hist=False,
             color='r', kde_kws={"shade": True})
hc_dist =sns.distplot(ls_obs_data.loc[ls_obs_data.condition == 'hc', 'rt'], hist=False 
            , kde_kws={"shade": True})
plt.ylabel('Density')
plt.xlabel('Reaction Time (s)')
plt.tight_layout()
sns.despine(left=True)
plt.show()

hv_dist = sns.distplot(ls_obs_data.loc[ls_obs_data.condition == 'hv', 'rt'], hist=False,
            kde_kws={"shade": True})
lv_dist = sns.distplot(ls_obs_data.loc[ls_obs_data.condition == 'lv', 'rt'], hist=False, kde_kws={"shade": True})
plt.ylabel('Density')
plt.xlabel('Reaction Time (s)')
plt.tight_layout()
sns.despine(left=True)
plt.show()

hv_dist.text(0.2, 4.5, "test", horizontalalignment='left', size='medium', color='black', weight='semibold')
```
