```python
confounds_fn = '/lab_data/coaxlab/Projects/loki_1/fmriprep_BOLD_data/sub-*/ses-*/func/*confounds*.tsv'
```

```python
confound_files = glob.glob(confounds_fn)
```

```python
all_deriv_quad_vars_df = pd.read_csv('/lab_data/coaxlab/Projects/loki_1/fmriprep_BOLD_data/sub-860/ses-03/func/sub-860_ses-03_task-lokicat_run-1_desc-confounds_timeseries.tsv', sep='\t')
```

```python
a_comp_cor_cols = [col for col in all_deriv_quad_vars_df if col.startswith('a_comp_cor')]
```

```python
a_comp_cor_cols_use = a_comp_cor_cols[:6] 
# new version of fmriprep only includes first 6 princ. components of anatomically derived noise signals 
```

```python
confound_dfs = []

for f in confound_files:
    temp_df = pd.read_csv(f, sep='\t', 
                          usecols=['trans_x', 'trans_y', 'trans_z', 'rot_x', 
                                    'rot_y', 'rot_z'] + a_comp_cor_cols_use) 
    
    temp_df['sub'] = int(f.split('sub-')[1][:3])
    temp_df['session'] = int(f.split('ses-')[1][:2])
    temp_df['slice'] = range(len(temp_df))

    if temp_df['sub'].unique()[0] == 860:
        temp_df['run'] = int(f.split('run-')[1][:1])
    else:
        temp_df['run'] = int(f.split('run-')[1][:2])

    print(f)
    print(temp_df.run.unique())
    
    confound_dfs.append(temp_df)
```

```python
confound_df = pd.concat(confound_dfs, sort=True)
```

# Key confounds
## Estimated head-motion parameters (the 6 rigid-body motion parameters (3 translations and 3 rotations) estimated relative to a reference image)):
* trans_x
* trans_y
* trans_z
* rot_x
* rot_y
* rot_z 

## Global signals:
* csf: the average signal within the anatomically-derived eroded CSF mask
* white_matter:  the average signal within the anatomically-derived eroded WM masks
* global_signal: the average signal within the brain mask
