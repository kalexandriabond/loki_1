# subcortical parcellation 

```python
from nilearn import datasets

dataset = datasets.fetch_atlas_pauli_2017() # get probabilistic subcortical atlas 
atlas_filename = dataset.maps # filename for the atlas
labels = dataset.labels # region labels

n_nodes = len(labels)
```

```python
plotting.plot_prob_atlas(atlas_filename, title='', 
                         bg_img=bold_ref_fn, display_mode='ortho', draw_cross=False, 
                        colorbar=False) # only include striatal regions?
plt.savefig(os.path.join(fig_path, 'ortho_prob_subcort_atlas.pdf'))
```

```python
plotting.plot_prob_atlas(atlas_filename, title='', 
                         bg_img=bold_ref_fn, display_mode='z', draw_cross=False, 
                        colorbar=False) 
plt.savefig(os.path.join(fig_path, 'axial_prob_subcort_atlas.pdf'))
```
