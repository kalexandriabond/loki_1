```python
iti_min = 2.5
iti_max = 8
rate_param = 1
```

```python
from scipy import stats
import numpy
```

```python
n_trials = 400
```

```python
lower, upper, scale = iti_min, iti_max, rate_param
X = stats.truncexpon(b=(upper-lower)/scale, loc=lower, scale=scale)
iti_data = X.rvs(n_trials)
```

```python
iti_data.min(), iti_data.max()
```

```python
fig, ax = plt.subplots()
ax.hist(iti_data, normed=False)
plt.show()
```
